```python
class Backtester:
    def __init__(self, initial_capital=100000.0, commission_pct=0.0):
        self.initial_capital = initial_capital
        self.commission_pct = commission_pct
        self.cash = initial_capital
        self.positions = {}
        self.portfolio_value = []
        self.trades = []
        self.strategy_func = None
    
    def load_data(self, data):
        self.data = data
    
    def set_strategy(self, strategy_func):
        self.strategy_func = strategy_func
    
    def execute_trades(self):
        if self.strategy_func == None:
            print(f'No strategy yet defined, please do this first')
            return
        
        self.signals = self.strategy_func(self.data)

        for date, signal in signals.items():
            price = self.data.loc[date]['Close']

            if signal == 'BUY' and self.cash > price:
                shares = int(self.cash / price)
                cost = shares * price * (1 + self.commission_pct)
                self.cash -= cost
                self.positions[date] = shares
                
                # Record trade details
                self.trades.append({
                    'Date': date,
                    'Type': 'BUY',
                    'Shares': shares,
                    'Price': price,
                    'Cost': cost
                })

            elif signal == 'SELL' and date in self.positions:
                shares = self.positions.pop(date)
                revenue = shares * price * (1 - self.commission_pct)
                self.cash += revenue
                
                # Record trade details
                self.trades.append({
                    'Date': date,
                    'Type': 'SELL',
                    'Shares': shares,
                    'Price': price,
                    'Revenue': revenue
                })

            portfolio_value = self.cash + sum(shares * price for shares in self.positions.values())
            
            self.portfolio_value.append(portfolio_value)

    def evaluate_performance(self):

        returns = pd.Series(self.portfolio_value).pct_change().dropna()

        print(returns)

        # total_return = (self.portfolio_value[-1] - self.initial_capital) / self.initial_capital
        # sharpe_ratio = returns.mean() / returns.std() * np.sqrt(252) # assuming daily returns

        #return {
            # "Total Return": total_return,
            # "Sharpe Ratio": sharpe_ratio,
            # "Final Portfolio Value": self.portfolio_value[-1],
            # "Trades": pd.DataFrame(self.trades) 
        # }
```

```python
# TODO change so it returns a dataframe with date and signal
def sma_strategy(data):
    short_window = 10
    long_window = 30

    data['SMA_Short'] = data['Close'].rolling(window=short_window).mean()
    data['SMA_Long'] = data['Close'].rolling(window=long_window).mean()

    signals = {}

    for i in range(len(data)):
        if i < long_window:
            continue
        
        date = data.index[i]
        if data['SMA_Short'][i] > data['SMA_Long'][i]:
            signals[date] = 'BUY'
        elif data['SMA_Short'][i] < data['SMA_Long'][i]:
            signals[date] = 'SELL'

    print(signals)
        
    return signals
```

```python
backtester = Backtester()
backtester.load_data(data=data)
backtester.set_strategy(sma_strategy)
backtester.execute_trades()
```

```python
results = backtester.evaluate_performance()
print(results)
```
